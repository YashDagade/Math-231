```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def die_rool():
    return random.randint(1, 6)
```

```python
def sum_of_n_dies(n):
    return sum([die_rool() for i in range(n)])

print(sum_of_n_dies(2))
```

```python
def plot_histogram_optimized(no_of_trials, no_of_dies):
    # Generate all random numbers at once (no_of_trials x no_of_dies matrix)
    rolls = np.random.randint(1, 7, size=(no_of_trials, no_of_dies))
    # Sum across rows to get sum for each trial
    results = np.sum(rolls, axis=1)
    
    # Calculate number of possible unique values
    min_sum = no_of_dies  # minimum possible sum (all 1s)
    max_sum = no_of_dies * 6  # maximum possible sum (all 6s)
    num_possible_values = max_sum - min_sum + 1
    
    plt.hist(results, bins=num_possible_values, 
             range=(min_sum - 0.5, max_sum + 0.5), 
             density=True)
    plt.show()

# Run with 100,000 trials, 2 dice
plot_histogram_optimized(1_000_000, 50)
```
