```python
# visualizing central limit thereorm

import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu = 0.5  # Probability of success in Bernoulli trials
N_values = [5, 10, 25, 100, 1000, 10000]  # Different sequence lengths
num_sequences = 10000  # Number of sequences to generate

# Set up the subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Iterate over different N values
for i, N in enumerate(N_values):
    # Generate sequences of Bernoulli random variables
    sequences = np.random.binomial(1, mu, (num_sequences, N))
    
    # Compute the mean m for each sequence
    means = np.mean(sequences, axis=1)
    
    # Plot the histogram of m
    ax = axs[i//3, i%3]  # Select subplot
    ax.hist(means, bins=30, density=True, alpha=0.7, color='blue')
    ax.set_title(f'N = {N}')
    ax.set_xlabel('Mean (m)')
    ax.set_ylabel('Density')
    ax.set_xlim(0, 1)
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Simulation parameters
n = 10000  # Number of draws
numbers = list(range(50))  # The set {0, 1, ..., 49}

# Initialize counters
divisible_by_7 = 0
divisible_by_7_and_3 = 0

# Simulate the random draws and count frequencies
for _ in range(n):
    num = rnd.choice(numbers)
    if num % 7 == 0:
        divisible_by_7 += 1
    if num % 7 == 0 and num % 3 == 0:
        divisible_by_7_and_3 += 1

# Calculate probabilities
prob_divisible_by_7 = divisible_by_7 / n
prob_divisible_by_7_and_3 = divisible_by_7_and_3 / n

# Theoretical probabilities
theoretical_prob_7 = len([x for x in numbers if x % 7 == 0]) / len(numbers)
theoretical_prob_7_and_3 = len([x for x in numbers if x % 7 == 0 and x % 3 == 0]) / len(numbers)

# Plotting the results
labels = ['Divisible by 7', 'Divisible by 7 and 3']
observed = [prob_divisible_by_7, prob_divisible_by_7_and_3]
theoretical = [theoretical_prob_7, theoretical_prob_7_and_3]

x = range(len(labels))

plt.bar(x, observed, width=0.4, label='Observed', align='center')
plt.bar(x, theoretical, width=0.4, label='Theoretical', align='edge')
plt.xticks(x, labels)
plt.ylabel('Probability')
plt.title('Observed vs Theoretical Probabilities')
plt.legend()
plt.show()

print(f'Observed probability of divisible by 7: {prob_divisible_by_7:.4f}')
print(f'Theoretical probability of divisible by 7: {theoretical_prob_7:.4f}')

print(f'\n \nObserved probability of divisible by 7 and 3: {prob_divisible_by_7_and_3:.4f}')
print(f'Theoretical probability of divisible by 7 and 3: {theoretical_prob_7_and_3:.4f}')
```

### Evaluate the following infinite sums:

1. $ \sum_{k=1}^{\infty} \left( \frac{1}{4} \right)^k $

This is an infinite geometric series with first term $a = \frac{1}{4}$ and common ratio $r = \frac{1}{4}$.

The sum of an infinite geometric series is:

$S = \frac{a}{1 - r}, \text{ for } |r| < 1$

Substituting the values:

$S = \frac{\frac{1}{4}}{1 - \frac{1}{4}} = \frac{\frac{1}{4}}{\frac{3}{4}} = \frac{1}{3}$

2. $ \sum_{k=1}^{\infty} \frac{3^k}{k!} $

This is the exponential series for $e^3$ minus the $k = 0$ term. The full series for $e^x$ is:

$e^x = \sum_{k=0}^{\infty} \frac{x^k}{k!}$

For $e^3$:

$e^3 = \sum_{k=0}^{\infty} \frac{3^k}{k!} = 1 + \frac{3^1}{1!} + \frac{3^2}{2!} + \frac{3^3}{3!} + \cdots$

Subtracting the first term:

$e^3 - 1 = \sum_{k=1}^{\infty} \frac{3^k}{k!}$

Calculating the first few terms:

$\frac{3^1}{1!} = 3$, $\frac{3^2}{2!} = 4.5$, $\frac{3^3}{3!} = 4.5$, $\frac{3^4}{4!} = 3.375$, $\frac{3^5}{5!} = 2.025$

Summing these terms:

$e^3 - 1 \approx 3 + 4.5 + 4.5 + 3.375 + 2.025 + \cdots = 19.08554$

### Finding the Area of the Region

We want to calculate the area of the region where $x$ and $y$ are between 0 and 1 and $y \geq x^2$. This corresponds to the region under the curve $y = x^2$ from $x = 0$ to $x = 1$.

To find the area, we compute the following definite integral:

$\text{Area} = \int_0^1 (1 - x^2) \, dx$

This integral represents the area between the line $y = 1$ and the curve $y = x^2$.

### Step-by-Step Calculation:

1. The integral of $1$ is $x$.
2. The integral of $x^2$ is $\frac{x^3}{3}$.

Thus, the area is:


$\text{Area} = \left[ x - \frac{x^3}{3} \right]_0^1 = \left( 1 - \frac{1^3}{3} \right) - \left( 0 - \frac{0^3}{3} \right) = 1 - \frac{1}{3} = \frac{2}{3}$

So, the area of the region is:


$ \frac{2}{3}$
